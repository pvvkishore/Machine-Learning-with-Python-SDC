<h2 style="margin-bottom:0.25rem;">Dataset: Forest CoverType (Covertype)</h2>
<ul>
  <li><b>Task:</b> Multiclass classification with <b>7 cover-type classes</b>.</li>
  <li><b>Input:</b> Tabular features (continuous + many binary indicators such as wilderness/soil types).</li>
  <li><b>Why ‚Äútough‚Äù:</b> Large dataset, nonlinear patterns, and class confusions that stress linear models.</li>
</ul>

<h3 style="margin-bottom:0.25rem;">Where the dataset can be inducted (loaded) from</h3>
<ul>
  <li><b>scikit-learn loader (recommended):</b> <code>sklearn.datasets.fetch_covtype</code> (auto-download + load)</li>
  <li><b>UCI repository:</b> Covertype dataset download files</li>
  <li><b>OpenML:</b> Hosted version of Covertype for easy programmatic retrieval</li>
</ul>

<hr style="border: 1px solid #ddd;" />

<h2 style="margin-bottom:0.25rem;">Evaluation Protocol (Fair Comparison)</h2>
<ul>
  <li><b>Same preprocessing</b> per model family (e.g., scaling for LR/kNN/SVM; no scaling required for RF).</li>
  <li><b>Same split & CV strategy:</b> Stratified train/test split + Stratified K-Fold CV on training only.</li>
  <li><b>Same tuning budget:</b> Comparable grid/random search effort for each model.</li>
  <li><b>Primary selection metric:</b> <span style="color:#006400;"><b>Macro F1</b></span> (treats classes equally).</li>
</ul>

<hr style="border: 1px solid #ddd;" />

<h2 style="color:#d00000; margin-bottom:0.25rem;">üî¥ Critical Failure Point (Where a model stops being ‚Äúgood‚Äù)</h2>
<div style="border-left: 6px solid #d00000; padding: 0.6rem 0.8rem; background: #fff3f3;">
  <p style="margin:0.1rem 0;">
    A model is considered to <b style="color:#d00000;">fail</b> when it hits either:
  </p>
  <ol style="margin:0.35rem 0 0.1rem 1.2rem;">
    <li><b style="color:#d00000;">Quality failure:</b> unacceptable predictive performance, especially on minority/hard classes.</li>
    <li><b style="color:#d00000;">Practical failure:</b> training/inference becomes infeasible at dataset scale (time/memory).</li>
  </ol>
</div>

<h3 style="margin-bottom:0.25rem;">Define explicit failure criteria (examples)</h3>
<ul>
  <li><b style="color:#d00000;">Macro F1 below threshold</b> (choose a target, e.g., 0.60), or</li>
  <li><b style="color:#d00000;">Any class recall below threshold</b> (e.g., &lt; 0.40 for at least one class), or</li>
  <li><b style="color:#d00000;">High instability</b>: large CV variance (mean ¬± std indicates unreliable performance).</li>
</ul>

<h3 style="margin-bottom:0.25rem;">How to locate the failure ‚Äúcritical point‚Äù (diagnostics)</h3>
<ul>
  <li><b>Learning curve stress test:</b> performance vs training-set fraction (5%, 10%, 25%, 50%, 100%).  
      <br/><span style="color:#d00000;"><b>Critical point</b></span> = where performance plateaus too low (underfitting) or degrades.</li>
  <li><b>Minority-class stress test:</b> per-class recall/F1 vs class frequency.  
      <br/><span style="color:#d00000;"><b>Critical point</b></span> = where rare-class recall collapses.</li>
  <li><b>Distribution-shift split (optional but revealing):</b> hold out a wilderness area or subset of soil types.  
      <br/><span style="color:#d00000;"><b>Critical point</b></span> = sharp drop in Macro F1 / per-class recalls under shift.</li>
  <li><b>Feasibility test:</b> record time/memory vs dataset size.  
      <br/><span style="color:#d00000;"><b>Critical point</b></span> = runtime/memory exceeds acceptable limits.</li>
</ul>

<hr style="border: 1px solid #ddd;" />

<h2 style="color:#006400; margin-bottom:0.25rem;">üü© Final Results to Report (What will decide suitability)</h2>
<ul>
  <li><b style="color:#006400;">Model ranking table:</b> Macro F1 (primary), Balanced Accuracy, Accuracy, Train time, Predict time.</li>
  <li><b style="color:#006400;">Per-class table:</b> Precision/Recall/F1 for each of 7 classes + worst-case recall.</li>
  <li><b style="color:#006400;">Confusion matrix:</b> identify systematic confusions between cover types.</li>
  <li><b style="color:#006400;">Stability:</b> CV mean ¬± std (Macro F1) to judge reliability.</li>
  <li><b style="color:#006400;">Failure report:</b> the detected critical point(s) for each model (quality + feasibility).</li>
</ul>

<h2 style="margin-bottom:0.25rem;">Decision Rule (How we pick the ‚Äúmost suitable‚Äù model)</h2>
<ol>
  <li>Pick the model with the <b style="color:#006400;">highest Macro F1</b> on held-out test (or outer CV if using nested CV).</li>
  <li>If close, prefer higher <b style="color:#006400;">minimum per-class recall</b> and lower CV variance (more robust).</li>
  <li>If still close, prefer better feasibility: lower training/prediction time and memory use.</li>
</ol>

```python
# Step 1: Imports + basic settings

import numpy as np
import pandas as pd
import time
import warnings

from sklearn.datasets import fetch_covtype
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, learning_curve
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score,
    balanced_accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix
)

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

warnings.filterwarnings("ignore")

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

pd.set_option("display.max_colwidth", 120)
pd.set_option("display.width", 120)
```

```python
# Step 2: Load dataset + quick inspection + validation + train/test split

# 1) Load Covertype (try as_frame=True; fallback if unavailable)
try:
    cov = fetch_covtype(as_frame=True)
    X = cov.data.copy()
    y = cov.target.copy()
except TypeError:
    cov = fetch_covtype()
    X = pd.DataFrame(cov.data, columns=cov.feature_names)
    y = pd.Series(cov.target, name="Cover_Type")

# 2) Make labels 0..6 (original is 1..7)
y = (y.astype(int) - 1).astype(int)

print("Dataset loaded: Forest CoverType (Covertype)")
print(f"X shape: {X.shape} | y shape: {y.shape}")
print(f"Number of classes: {y.nunique()} (labels: {sorted(y.unique())})")

# 3) Display a portion of the dataset (top rows + target)
preview = X.head(5).copy()
preview["target"] = y.head(5).values
display(preview)

# 4) Basic sanity checks: format suitable for ML training
print("\n--- Data sanity checks ---")
print("Any missing values in X?", X.isna().any().any())
print("Total missing values in X:", int(X.isna().sum().sum()))
print("X dtypes summary:")
display(X.dtypes.value_counts())

# Confirm numeric features only (expected for this dataset)
non_numeric_cols = X.columns[~X.apply(lambda s: np.issubdtype(s.dtype, np.number))].tolist()
print("Non-numeric columns:", non_numeric_cols if non_numeric_cols else "None ‚úÖ")

# Show feature name order (to confirm consistent ordering)
print("\nFirst 10 feature columns:", list(X.columns[:10]))
print("Last 5 feature columns:", list(X.columns[-5:]))

# 5) Show class distribution (overall)
print("\nOverall class distribution (proportion):")
display(y.value_counts(normalize=True).sort_index().rename("proportion"))

# 6) Train/test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print("\n--- After split ---")
print(f"X_train: {X_train.shape} | y_train: {y_train.shape}")
print(f"X_test : {X_test.shape}  | y_test : {y_test.shape}")

# 7) Verify split integrity + stratification quality
print("\nTrain class distribution (proportion):")
display(y_train.value_counts(normalize=True).sort_index().rename("train_prop"))

print("Test class distribution (proportion):")
display(y_test.value_counts(normalize=True).sort_index().rename("test_prop"))

# 8) Ensure training/testing are aligned & usable
assert list(X_train.columns) == list(X_test.columns), "Column mismatch between train/test!"
assert X_train.shape[0] == y_train.shape[0] and X_test.shape[0] == y_test.shape[0], "Row/label mismatch!"
assert set(X_train.index).isdisjoint(set(X_test.index)), "Train/Test overlap detected!"
print("\nFormat check: train/test columns aligned, indices disjoint, shapes consistent ‚úÖ")
```

```python
# Step 2 (updated): Load Covertype + preview + label-name mapping + validation + train/test split

# 1) Load Covertype (try as_frame=True; fallback if unavailable)
try:
    cov = fetch_covtype(as_frame=True)
    X = cov.data.copy()
    y_raw = cov.target.copy()  # original labels 1..7
except TypeError:
    cov = fetch_covtype()
    X = pd.DataFrame(cov.data, columns=cov.feature_names)
    y_raw = pd.Series(cov.target, name="Cover_Type")  # original labels 1..7

# 2) Create label mapping (original 1..7) -> cover type name (UCI/KDD mapping)
cover_type_map_1to7 = {
    1: "Spruce/Fir",
    2: "Lodgepole Pine",
    3: "Ponderosa Pine",
    4: "Cottonwood/Willow",
    5: "Aspen",
    6: "Douglas-fir",
    7: "Krummholz",
}

# Convert to 0..6 for training (and build matching 0..6 name map)
y = (y_raw.astype(int) - 1).astype(int)
cover_type_map_0to6 = {k - 1: v for k, v in cover_type_map_1to7.items()}

print("Dataset loaded: Forest CoverType (Covertype)")
print(f"X shape: {X.shape} | y shape: {y.shape}")
print(f"Number of classes: {y.nunique()} (labels: {sorted(y.unique())})")

# Display label mapping
label_df = pd.DataFrame({
    "label_0to6": list(cover_type_map_0to6.keys()),
    "cover_type_name": list(cover_type_map_0to6.values())
}).sort_values("label_0to6").reset_index(drop=True)

print("\nLabel mapping (used in this notebook):")
display(label_df)

# 3) Display a portion of the dataset (top rows + target label + target name)
preview = X.head(5).copy()
preview["target_label_0to6"] = y.head(5).values
preview["target_name"] = preview["target_label_0to6"].map(cover_type_map_0to6)
display(preview)

# 4) Basic sanity checks: format suitable for ML training
print("\n--- Data sanity checks ---")
print("Any missing values in X?", X.isna().any().any())
print("Total missing values in X:", int(X.isna().sum().sum()))
print("X dtypes summary:")
display(X.dtypes.value_counts())

# Confirm numeric features only (expected for this dataset)
non_numeric_cols = X.columns[~X.apply(lambda s: np.issubdtype(s.dtype, np.number))].tolist()
print("Non-numeric columns:", non_numeric_cols if non_numeric_cols else "None ‚úÖ")

# Show feature name order (to confirm consistent ordering)
print("\nFirst 10 feature columns:", list(X.columns[:10]))
print("Last 5 feature columns:", list(X.columns[-5:]))

# 5) Show class distribution (overall) with names
class_counts = y.value_counts().sort_index()
class_props = y.value_counts(normalize=True).sort_index()

dist_df = pd.DataFrame({
    "count": class_counts,
    "proportion": class_props,
    "cover_type_name": [cover_type_map_0to6[i] for i in class_counts.index]
})
print("\nOverall class distribution:")
display(dist_df)

# 6) Train/test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=RANDOM_STATE, stratify=y
)

print("\n--- After split ---")
print(f"X_train: {X_train.shape} | y_train: {y_train.shape}")
print(f"X_test : {X_test.shape}  | y_test : {y_test.shape}")

# 7) Verify stratification quality (train vs test proportions) with names
train_props = y_train.value_counts(normalize=True).sort_index()
test_props = y_test.value_counts(normalize=True).sort_index()

strat_df = pd.DataFrame({
    "train_prop": train_props,
    "test_prop": test_props,
    "cover_type_name": [cover_type_map_0to6[i] for i in train_props.index]
})
print("\nStratification check (proportions):")
display(strat_df)

# 8) Ensure training/testing are aligned & usable
assert list(X_train.columns) == list(X_test.columns), "Column mismatch between train/test!"
assert X_train.shape[0] == y_train.shape[0] and X_test.shape[0] == y_test.shape[0], "Row/label mismatch!"
assert set(X_train.index).isdisjoint(set(X_test.index)), "Train/Test overlap detected!"
print("\nFormat check: train/test columns aligned, indices disjoint, shapes consistent ‚úÖ")
```

```python
# Step 3: Evaluation framework (metrics + CV) + model definitions (pipelines + grids)
#         Also sets up label names so later reports/confusion matrices show cover-type names.

from sklearn.metrics import make_scorer

# 0) Label names (expects cover_type_map_0to6 created in Step 2)
# If Step 2 ran, this should exist already. We'll defensively rebuild if not.
if "cover_type_map_0to6" not in globals():
    cover_type_map_0to6 = {
        0: "Spruce/Fir",
        1: "Lodgepole Pine",
        2: "Ponderosa Pine",
        3: "Cottonwood/Willow",
        4: "Aspen",
        5: "Douglas-fir",
        6: "Krummholz",
    }

CLASS_NAMES = [cover_type_map_0to6[i] for i in sorted(cover_type_map_0to6.keys())]
print("Class names (label order 0..6):")
display(pd.DataFrame({"label_0to6": range(len(CLASS_NAMES)), "cover_type_name": CLASS_NAMES}))

# 1) Metrics
PRIMARY_METRIC_NAME = "f1_macro"  # primary: treats all classes equally

SCORERS = {
    "accuracy": make_scorer(accuracy_score),
    "balanced_accuracy": make_scorer(balanced_accuracy_score),
    "f1_macro": make_scorer(f1_score, average="macro"),
}

print("\nPrimary metric:", PRIMARY_METRIC_NAME)
print("Metrics to compute:", list(SCORERS.keys()))

# 2) Cross-validation strategy (fair & stratified)
CV = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
print("\nCV strategy:", CV)

# 3) Models + hyperparameter grids
# Note: Scaling is required for LR, kNN, and SVM; Random Forest doesn't need scaling.

MODELS = {
    "LogisticRegression": {
        "estimator": Pipeline([
            ("scaler", StandardScaler()),
            ("clf", LogisticRegression(
                max_iter=3000,
                n_jobs=-1,
                class_weight="balanced",
                solver="saga"
            ))
        ]),
        "param_grid": {
            "clf__C": [0.1, 1.0, 10.0],
            "clf__penalty": ["l2"],
        }
    },

    "kNN": {
        "estimator": Pipeline([
            ("scaler", StandardScaler()),
            ("clf", KNeighborsClassifier())
        ]),
        "param_grid": {
            "clf__n_neighbors": [5, 15, 35],
            "clf__weights": ["uniform", "distance"],
            "clf__p": [1, 2],  # Manhattan vs Euclidean
        }
    },

    "RandomForest": {
        "estimator": RandomForestClassifier(
            random_state=RANDOM_STATE,
            n_jobs=-1,
            class_weight="balanced_subsample"
        ),
        "param_grid": {
            "n_estimators": [300],
            "max_depth": [None, 20, 40],
            "min_samples_split": [2, 10],
            "min_samples_leaf": [1, 5],
            "max_features": ["sqrt", 0.5],
        }
    },

    "SVM_RBF": {
        "estimator": Pipeline([
            ("scaler", StandardScaler()),
            ("clf", SVC(
                kernel="rbf",
                class_weight="balanced"
            ))
        ]),
        "param_grid": {
            "clf__C": [1.0, 10.0],
            "clf__gamma": ["scale", 0.01],
        }
    }
}

# 4) Quick summary so you can verify everything is set before running training
summary_rows = []
for name, cfg in MODELS.items():
    grid = cfg["param_grid"]
    # total combinations in the grid
    total_combos = 1
    for v in grid.values():
        total_combos *= len(v)
    summary_rows.append({
        "model": name,
        "uses_scaling": isinstance(cfg["estimator"], Pipeline),
        "grid_params": list(grid.keys()),
        "grid_total_combinations": total_combos
    })

print("\nModel setup summary:")
display(pd.DataFrame(summary_rows))

print("Step 3 complete ‚úÖ (No training has been run yet.)")
```

```python
# Step 4: Run GridSearchCV for each model + evaluate on test set (collect predictions for later analysis)

from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

# Storage for results and artifacts
results = []
best_estimators = {}
test_predictions = {}

# For reproducible CV scoring and fair comparison
SCORING = PRIMARY_METRIC_NAME  # "f1_macro"

print(f"Running GridSearchCV for {len(MODELS)} models with scoring='{SCORING}' and {CV.get_n_splits()}-fold CV...\n")

for model_name, cfg in MODELS.items():
    estimator = cfg["estimator"]
    param_grid = cfg["param_grid"]

    print(f"=== {model_name} ===")
    print("Grid parameters:", param_grid)

    gs = GridSearchCV(
        estimator=estimator,
        param_grid=param_grid,
        scoring=SCORING,
        cv=CV,
        n_jobs=-1,
        verbose=0,
        refit=True,
        return_train_score=False
    )

    t0 = time.time()
    gs.fit(X_train, y_train)
    train_seconds = time.time() - t0

    best_estimators[model_name] = gs.best_estimator_

    # Test evaluation (same metrics for all models)
    y_pred = gs.best_estimator_.predict(X_test)
    test_predictions[model_name] = y_pred

    test_acc = accuracy_score(y_test, y_pred)
    test_bal_acc = balanced_accuracy_score(y_test, y_pred)
    test_f1_macro = f1_score(y_test, y_pred, average="macro")

    results.append({
        "model": model_name,
        "best_cv_f1_macro": gs.best_score_,
        "test_accuracy": test_acc,
        "test_balanced_accuracy": test_bal_acc,
        "test_f1_macro": test_f1_macro,
        "train_time_sec": train_seconds,
        "best_params": gs.best_params_,
        "n_train": X_train.shape[0],
        "n_test": X_test.shape[0],
    })

    print(f"Best CV f1_macro: {gs.best_score_:.4f}")
    print(f"Test  Accuracy:   {test_acc:.4f}")
    print(f"Test  BalAcc:     {test_bal_acc:.4f}")
    print(f"Test  F1-macro:   {test_f1_macro:.4f}")
    print(f"Train time (sec): {train_seconds:.1f}")
    print("Best params:", gs.best_params_)
    print()

# Results table (rank by primary metric on test set)
results_df = pd.DataFrame(results).sort_values("test_f1_macro", ascending=False).reset_index(drop=True)

print("=== Summary (ranked by test_f1_macro) ===")
display(results_df[[
    "model",
    "best_cv_f1_macro",
    "test_f1_macro",
    "test_balanced_accuracy",
    "test_accuracy",
    "train_time_sec",
    "best_params"
]])

# Quick winner printout
best_model = results_df.loc[0, "model"]
print(f"Current best by test_f1_macro: {best_model}")
```

```python
# Step 5: Detailed diagnostics per model (named classes)
# - Classification report (per-class precision/recall/F1)
# - Confusion matrix
# - "Failure spotlight": weakest classes by recall and F1 (critical failure view)

from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Helper: map numeric labels (0..6) to names consistently
label_order = list(range(len(CLASS_NAMES)))  # [0,1,2,3,4,5,6]
name_order = [cover_type_map_0to6[i] for i in label_order]

def compute_failure_spotlight(y_true, y_pred, class_names):
    """
    Returns a DataFrame with per-class precision/recall/f1 and highlights weakest classes.
    """
    report = classification_report(
        y_true, y_pred,
        labels=label_order,
        target_names=class_names,
        output_dict=True,
        zero_division=0
    )
    rows = []
    for cname in class_names:
        rows.append({
            "class": cname,
            "precision": report[cname]["precision"],
            "recall": report[cname]["recall"],
            "f1": report[cname]["f1-score"],
            "support": report[cname]["support"],
        })
    df = pd.DataFrame(rows).sort_values("recall").reset_index(drop=True)
    return df

# Run diagnostics for each model
for model_name in results_df["model"].tolist():  # use ranking order
    y_pred = test_predictions[model_name]

    print("\n" + "="*90)
    print(f"MODEL: {model_name}")
    print("="*90)

    # 1) Overall metrics (already computed, re-print for convenience)
    acc = accuracy_score(y_test, y_pred)
    bal_acc = balanced_accuracy_score(y_test, y_pred)
    f1m = f1_score(y_test, y_pred, average="macro")
    print(f"Test Accuracy: {acc:.4f} | Balanced Acc: {bal_acc:.4f} | Macro F1: {f1m:.4f}")

    # 2) Classification report with readable class names
    print("\n--- Classification Report (Test) ---")
    print(classification_report(
        y_test, y_pred,
        labels=label_order,
        target_names=name_order,
        digits=4,
        zero_division=0
    ))

    # 3) Confusion matrix (named axes)
    cm = confusion_matrix(y_test, y_pred, labels=label_order)
    fig, ax = plt.subplots(figsize=(8, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=name_order)
    disp.plot(ax=ax, xticks_rotation=45, values_format="d", colorbar=False)
    ax.set_title(f"Confusion Matrix (Test) ‚Äî {model_name}")
    plt.tight_layout()
    plt.show()

    # 4) Failure spotlight: weakest classes (critical failure view)
    spot = compute_failure_spotlight(y_test, y_pred, name_order)

    print("--- Failure Spotlight (sorted by lowest recall) ---")
    display(spot)

    worst = spot.iloc[0]
    print(f"‚û°Ô∏è Worst class by recall: '{worst['class']}' | recall={worst['recall']:.4f} | f1={worst['f1']:.4f} | support={int(worst['support'])}")

    # Show top 3 weakest (often more informative than only 1)
    print("‚û°Ô∏è Three weakest classes by recall:")
    display(spot.head(3)[["class", "recall", "f1", "support"]])

print("\nStep 5 complete ‚úÖ")
```
